```python
from transformers import pipeline

pipe = pipeline("image-classification", model="chriamue/bird-species-classifier")
pipe("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/hub/parrots.png")
```

```python
from transformers import pipeline

pipe = pipeline("image-classification",
                model="chriamue/bird-species-classifier")

rock_dove_url = 'https://images.pexels.com/photos/1406506/pexels-photo-1406506.jpeg'
american_robin_url = 'https://images.pexels.com/photos/416179/pexels-photo-416179.jpeg'
res = pipe(american_robin_url)

print(res)
```
